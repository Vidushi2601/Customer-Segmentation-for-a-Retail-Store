## Reading the Cleaned File
the cleaned file 'Cleaned_Mall_Customers.csv' contains five parameters i.e. CustomerID,	Gender,	Age,	Annual Income (k$) and 	Spending Score (1-100)


Gender has been further numerically Encoded

```python
import pandas as pd

# Load the dataset
file_path = '/content/Cleaned_Mall_Customers .csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head(10)
```

```python
# Renaming columns for better readability
data.columns = ["CustomerID", "Gender", "Age", "AnnualIncome", "SpendingScore"]
```

# Descriptive statistics of the dataset

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualizations of age distribution, annual income distribution, and spending score distribution.

```python
#histplot-Age,Annual income,Spending score
# Visualizing distributions
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=30, kde=True, color='salmon')
plt.title('Age Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['AnnualIncome'], bins=30, kde=True, color='lightgreen')
plt.title('Annual Income Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['SpendingScore'], bins=30, kde=True)
plt.title('Spending Score Distribution')
plt.show()

```

## A scatter plot showing the relationship between annual income and spending score

```python
# Visualizing relationships
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='AnnualIncome', y='SpendingScore', hue='Gender')
plt.title('Income vs Spending Score')
plt.show()
```
